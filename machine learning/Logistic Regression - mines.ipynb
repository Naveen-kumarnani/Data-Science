```python
elist=dfmines.columns[:60]
```

```python
from scipy.stats import norm
```

```python
for e in elist:
    plt.subplot(1,2,1)
    plt.hist(dfmines[e],ec="r",bins=20)
    plt.subplot(1,2,2)
    sns.distplot(dfmines[e],fit=norm,hist=False)
    plt.show()
    print(" distributioin :",dfmines[e].skew())
    if dfmines[e].skew().round()==0:
        print("It is a Normal distribution ==> Mean std method")
        print(dfmines[e].describe())
        m=dfmines[e].mean()
        std=dfmines[e].std()
        lower=m-(3*std)
        upper=m+(3*std)
        print("lower limit :",lower)
        print("upper limit :",upper)
        dfmines[e]=dfmines[e].map(lambda d : m if d<lower else m if d>upper else d)
        print("Revalidation Errors ")
        print(dfmines[(dfmines[e]<lower) | (dfmines[e]>upper)])
    else:
        print("It is a skew distribution ==> IQR method")
        q1=dfmines[e].quantile(0.25)
        q3=dfmines[e].quantile(0.75)
        Iqr=q3-q1
        lower1=q1-(1.5*Iqr)
        upper1=q3+(1.5*Iqr)
        print("lower limit :",lower1)
        print("upper limit :",upper1)
        dfmines[e]=dfmines[e].map(lambda i : lower1 if i<lower1 else upper1 if i>upper1 else e )
        print("Revalidation Errors")
        print(dfmines[(dfmines[e]<lower1) | (dfmines[e]>upper1)])
        
```
