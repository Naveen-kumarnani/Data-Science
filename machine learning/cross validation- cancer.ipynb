```python
from scipy.stats import norm
for i in ilist:
    print(dfcancer[i])
    plt.subplot(1,2,1)
    sns.distplot(dfcancer[i],fit=norm)
    plt.subplot(1,2,2)
    plt.hist(dfcancer[i],ec="r",bins=20)
    plt.show()
    if dfcancer[i].skew().round()==0:
        print("It is a Normal distribution => Mean std method")
        print(dfcancer[i].describe())
        m=dfcancer[i].mean()
        s=dfcancer[i].std()
        lower=m-(3*s)
        upper=m+(3*s)
        print("lower limit :",lower)
        print("upper limit :",upper)
        dfcancer[i]=dfcancer[i].map(lambda t:m if t<lower else m if t>upper else t)
        print("Revalidation Errors")
        print(dfcancer[(dfcancer[i]<lower) | (dfcancer[i]>upper)])
    else:
        print(dfcancer[i])
        print("It is a skew distribution => IQR method")
        q1=dfcancer[i].quantile(0.25)
        q3=dfcancer[i].quantile(0.75)
        IQR=q3-q1
        lower1=q1-(1.5*IQR)
        upper1=q3+(1.5*IQR)
        print("lower limit :",lower1)
        print("upper limit :",upper1)
        dfcancer[i]=dfcancer[i].map(lambda t:lower1 if t<lower1 else upper1 if t>upper1 else t)
        print("Revalidation Errors")
        print(dfcancer[(dfcancer[i]<lower1) | (dfcancer[i]>upper1)])
    
        
        
        
        
        
```
