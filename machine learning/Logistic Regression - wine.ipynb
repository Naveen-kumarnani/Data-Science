### Error Detection

```python
plt.figure(figsize=(30,15))
sns.boxplot(data=dfwine)
```

```python
dfwine.columns
```

```python
len(dfwine.columns)
```

```python
dfwine.columns[:13]
```

```python
ilist=dfwine.columns[:13]
```

```python
from scipy.stats import norm
```

```python
for i in ilist:
    plt.subplot(1,2,1)
    plt.hist(dfwine[i],ec="r",bins=20)
    plt.subplot(1,2,2)
    sns.distplot(dfwine[i],fit=norm)
    plt.show()
    print(dfwine[i].skew())
    if dfwine[i].skew().round()==0:
        print("It is a Normal distribution ==> Mean std method")
        print(dfwine[i].describe())
        m=dfwine[i].mean()
        std=dfwine[i].std()
        lower=m-(3*std)
        upper=m+(3*std)
        print("lower limit :",lower)
        print("upper limit :",upper)
        dfwine[i]=dfwine[i].map(lambda d : m if d<lower else m if d>upper else d)
        print("Revalidation Errors ")
        print(dfwine[(dfwine[i]<lower) | (dfwine[i]>upper)])
    else:
        print("It is a skew distribution ==> IQR method")
        q1=dfwine[i].quantile(0.25)
        q3=dfwine[i].quantile(0.75)
        Iqr=q3-q1
        lower1=q1-(1.5*Iqr)
        upper1=q3+(1.5*Iqr)
        print("lower limit :",lower1)
        print("upper limit :",upper1)
        dfwine[i]=dfwine[i].map(lambda e : lower1 if e<lower1 else upper1 if e>upper1 else e )
        print("Revalidation Errors")
        print(dfwine[(dfwine[i]<lower1) | (dfwine[i]>upper1)])
        
```
