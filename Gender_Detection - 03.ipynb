```python
for i in path_male:
    img=cv2.imread(path+fl[1]+"/"+i)
    img=cv2.resize(img,(100,100))
    img_incol.append(img)
    img_depecol.append(1)
```

```python
len(img_depecol)
```

```python
len(img_incol)
```

```python
incol_ary=np.array(img_incol)
```

```python
depen_ary=np.array(img_depecol)
```

```python
depen_ary
```

```python
incol_ary
```

### Scale

```python
x=incol_ary/255.0
```

```python
y=depen_ary
```

```python
x.max()
```

```python
x.min()
```

### Spliting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x.shape,X_train.shape,X_test.shape
```

```python
y.shape,y_train.shape,y_test.shape
```

### Model Building

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Input,Flatten,Conv2D,MaxPooling2D,Dropout,Dense
```

```python
smodel=Sequential()
```

```python
smodel.add(Input(shape=(100,100,3)))
```

```python
smodel.add(Conv2D(filters=64,kernel_size=(3,3),activation="relu",strides=(1,1),padding="valid"))
smodel.add(MaxPooling2D(pool_size=(2,2),strides=2,padding="valid"))
smodel.add(Dropout(0.20))
```

```python
smodel.add(Conv2D(filters=32,kernel_size=(3,3),activation="relu",strides=(1,1),padding="valid"))
smodel.add(MaxPooling2D(pool_size=(2,2),strides=2,padding="valid"))
smodel.add(Dropout(0.20))
```

### Fully connected Layers / Dense

```python
smodel.add(Flatten())
```

```python
smodel.add(Dense(128,activation="relu"))
```

```python
smodel.add(Dense(64,activation="relu"))
```

```python
smodel.add(Dense(32,activation="relu"))
```

```python
smodel.add(Dense(16,activation="relu"))
```

```python
smodel.add(Dense(2,activation="softmax"))
```

```python
smodel.summary()
```

```python
smodel.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
smodel.fit(X_train,y_train,epochs=20)
```

```python
smodel.evaluate(X_test,y_test)
```
